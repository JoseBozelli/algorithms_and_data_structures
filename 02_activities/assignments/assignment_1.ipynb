````python
# Your answer here
from typing import List

def first_duplicate(nums: List[int]) -> int:
    seen = set()
    for item in nums:
        if item in seen:
            return item
        else:
            seen.add(item)
    return -1
````

# Explain why your solution works
**The solution provided works because as one loops through the list from left to right, the code check if the current number is already in a set. If it is, it will return that number. If it isn't in the set, it will add that number to the set and move to the next number. If by the end of the list no duplicates have been found, it returns -1.**

# Explain the problem’s time and space complexity
**In the solution provided, one iterates through the list, looking up, and inserting into a Python set. To interate throught the list the time complexity is O(n), that is, in the worst case the time to execute it will be linearly dependent on the input size (n). While on average looking up and inserting into a Python set the time complexity is O(1), i.e, the operation takes constant time regardless of the size of the input. Hence, we have O(n+1) in total, which makes the time complexity O(n).**
**Regarding the space complexity, in the worst case scenario when no duplicates are found, one will have to store all n elements into the set. Hence, the space complexity increases linearly with the input size (n) and, therefore the space complexity is O(n).**

# Explain the thinking to an alternative solution (no coding required, but a classmate reading this should be able to code it up based off your text)

**An alternative solution would be to use a nested loop. The first (outer) loop selects an element of the list (scanning from left to right), while the second (inner) loop iterates through the rest of the list looking for an element that is equal to the element selected in the first (outer) loop. When the condition is met, it will return that element. In the worst case scenario, when no duplicates are found, it will return -1.**
**In this alternative solution the time complexity is O(n^2). The reason is because in the worst case scenario, where no duplicates are found, one needs to iterate through all elements n*n (nested loop). On the other hand, the trade off comes in space complexity. As one don't need to store any data during duplicate search, the space complexity is O(1), i.e., operation takes constant time regardless of the size of the input.**
