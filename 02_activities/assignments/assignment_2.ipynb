#### Starter Code for Question 1

````python
from collections import deque

# Definition for a binary tree node.
class TreeNode(object):
  def __init__(self, val = 0, left = None, right = None):
    self.val = val
    self.left = left
    self.right = right

def is_duplicate(root: TreeNode) -> int:
  # Base case. If node is empty, stop looking.
  if root is None:
      return -1
  
  seen = set()  # Set tracks unique values in visited nodes values.

  queue = deque([root]) # Start a Breadth-first search (BFS) using queue. Hierarchical search, level-by-level.
  
  while queue:  # Searching using a queue.
    current = queue.popleft() # Establish current node.

    if current.val in seen: # Check current node. If value in set ("seen"), return its value. Otherwise add to set ("seen") for tracking.
      return current.val
    else:
      seen.add(current.val)
    
    if current.left is not None: # Add left children to the queue. 
      queue.append(current.left)

    if current.right is not None: # Add right children o the queue. Will process after left child(ren).
      queue.append(current.right)
  
  return -1
````

**Trace/walkthrough**
My partner used pair of characters in strings to describe open/close or sealed/unsealed states. Their code determine whether in the provided string there is a matching character of the same type and in correct order. For instance, given the input "BbTt", where "B" = open Burger bun, "b" = close Burger bun, "T" = open Taco shell, "t" = close taco shell, the output is True. That is, the open burger bun, "B", has its matching close burger bun "b" in the correct order (i.e., open and, then, close). The same holds for "T", open taco shell, with its matching "t", close taco shell, in correct order. Hence, the output is True as in the string in the input there are two pairs of characters that match (or are the same type) and in the correct order, i.e., open/close.


-   Copy the solution your partner wrote. 


````python
# Your answer here
def is_valid_brackets(s: str) -> bool:
    #create an empty list to store valid strings
    stack = [] 
    #define the valid pairs
    pairs  = {")": "(", "}": "{", "]": "["}  

#create a for loop:
    for char in s:
        if char in "({[":
            #append the first left characters
            stack.append(char)
        else:
            if not stack:
                return  False
            # POP  = remove the last item from the list AND hand that item back to you 
            # ugh suffered on this one. 
            # the whole idea was to remove the items (i.e., pop) out if they match . 
            # Once the stack is empty, it means that everything matched up 
            top  = stack.pop()
            if top != pairs[char]:
                return False
    #stack empty  = False, therefore not stack = True     
    return not stack    
````


-   Explain why their solution works in your own words.


**In their code chunck they used a for loop to iterate through the characters of the string. They used an empty list as an accumulator as they loop over the string. However, since they used .append() and .pop() to manipulate the data, it is more accurately to actually call the accumulator a stack. They used a Python dictionary to define character matching pairs (in this case they only defined type of brackets). As the code loop over the string one character at a time, they first check if the character is one of the possible opening (i.e., left) brackets. If it is an opening bracket, it appends (i.e., stores to track) to stack. If it is a closing (i.e., right) bracket (i.e., else block) and the stack is empty, it returns False because the bracket has no corresponding matching bracket. Otherwise, it "pops" (i.e., removes) the last (note: stacks follows the last-in-first-out order, making it ideal for this task) appended left bracket into the stack and stores it in the "top" variable. Then, it compares the current right bracket with the last left bracket by looking up the dictionary with matching pairs. If they match, it loop over. If they don't match it returns False. In the end, if the stack is empty it returns True, meaning every left bracket has its matching right bracket in order.**


-   Explain the problem’s time and space complexity in your own words.


**In the solution provided, one iterates through the string, looking up one character. To interate through the string the time complexity is O(n), that is, in the worst case the number of steps and, therefore, the time to execute it will be linearly dependent on the input size (n). Regarding the space complexity, in the worst case scenario when no closing brackets are found, one will have to store all n elements (in the input) into the stack. Hence, the space complexity also increases linearly with the input size (n) and, therefore the space complexity is O(n).**
